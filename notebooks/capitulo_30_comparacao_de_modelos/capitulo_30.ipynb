## O Grande Quadro

A tabela abaixo resume nossa jornada. Todos os números vêm da mesma validação cruzada de 5 *folds* sobre o conjunto de treino canônico (480 amostras, 30 *features*), com destaque para a métrica que sempre nos importou: o **recall da classe maligna**.

| Modelo | Acurácia (CV) | Recall Maligno | Interpretab. | Observação |
|---|---|---|---|---|
| Naive Bayes | 0,9437 | 0,9037 | Alta | Simples e rápido; boa linha de base. |

| **Ensemble de Votação** | **0,9812** | **0,9662** | Baixa | **Modelo final**: combina os três melhores. |

### Principais Lições

**Não existe almoço grátis.** Nenhum modelo venceu em todas as dimensões. O Ensemble deu a melhor acurácia e o melhor recall maligno, mas ao custo de complexidade e opacidade. A Regressão Logística foi quase tão boa e muito mais interpretável.

**A complexidade nem sempre vence.** A Regressão Logística, um dos modelos mais simples, superou ensembles como o Random Forest e o Gradient Boosting neste dataset. Lição prática: **sempre comece por uma linha de base simples** — muitas vezes ela já é boa o suficiente.

**O poder da combinação — quando há diversidade.** O Ensemble superou *todos* os membros individuais (0,9812 contra 0,9771 do melhor deles) justamente porque combinou modelos de famílias diferentes, cujos erros não são totalmente correlacionados. Não foi mágica: foi diversidade.

**Interpretabilidade vs. desempenho.** A jornada da Regressão Logística (transparente) ao Ensemble (opaco) espelha um *trade-off* clássico. Nós sacrificamos interpretabilidade por desempenho — e depois a "compramos de volta" com SHAP e LIME. Essa dinâmica está no coração de muitos projetos reais.
